```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Dataset .csv')
plt.figure(figsize=(10, 6))
plt.hist(df['Aggregate rating'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Aggregate rating'], fill=True, color='skyblue')
plt.title('Kernel Density Estimation of Ratings')
plt.xlabel('Rating')
plt.ylabel('Density')
plt.show()
```

```python
top_cuisines = df['Cuisines'].value_counts().head(10).index
average_ratings_cuisine = df[df['Cuisines'].isin(top_cuisines)].groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 8))
sns.barplot(x=average_ratings_cuisine.index, y=average_ratings_cuisine.values, palette='viridis')
plt.title('Average Ratings of Top 10 Cuisines')
plt.xlabel('Cuisine Type')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Has Table booking'], df['Has Online delivery'], c=df['Aggregate rating'], cmap='viridis')
plt.title('Scatter Plot of Features against Target Variable')
plt.xlabel('Has Table booking')
plt.ylabel('Has Online delivery')
plt.colorbar(label='Aggregate rating')
plt.show()
```
